### Phase 1: ä½è¤‡é›œåº¦é«˜æ”¶ç›Š (Month 1-2)

````python
# ğŸš€ Phase 1: åŸºç¤è‡ªå‹•åŒ–å·¥å…·

import os
import json
import re
from datetime import datetime
from pathlib import Path

class BasicAutomationTools:
    """
    Phase 1 åŸºç¤è‡ªå‹•åŒ–å·¥å…·é›†
    å°ˆæ³¨æ–¼ä½é¢¨éšªã€é«˜æ”¶ç›Šçš„è‡ªå‹•åŒ–
    """
    
    def __init__(self, knowledge_base_path="./knowledge_base"):
        self.kb_path = Path(knowledge_base_path)
        self.kb_path.mkdir(exist_ok=True)
        
        # å»ºç«‹æ¨™æº–ç›®éŒ„çµæ§‹
        for subdir in ['01_raw', '02_processed', '03_indexed', '04_metadata', '05_quality', '06_lineage']:
            (self.kb_path / subdir).mkdir(exist_ok=True)
    
    def auto_file_organization(self, source_dir):
        """
        1. è‡ªå‹•æ–‡ä»¶æ•´ç†ï¼šæ ¹æ“šæª”æ¡ˆé¡å‹å’Œå‘½åè¦å‰‡è‡ªå‹•åˆ†é¡
        """
        source_path = Path(source_dir)
        if not source_path.exists():
            print(f"âŒ ä¾†æºç›®éŒ„ä¸å­˜åœ¨: {source_dir}")
            return
        
        print("ğŸ—‚ï¸ åŸ·è¡Œè‡ªå‹•æ–‡ä»¶æ•´ç†...")
        
        # æ–‡ä»¶åˆ†é¡è¦å‰‡
        classification_rules = {
            'æœƒè­°': ['meeting', 'meeting_notes', 'æœƒè­°è¨˜éŒ„', 'æœƒè­°'],
            'å ±å‘Š': ['report', 'analysis', 'å ±å‘Š', 'åˆ†æ'],
            'è¦æ ¼': ['spec', 'specification', 'è¦æ ¼', 'è¦ç¯„'],
            'æ¸¬è©¦': ['test', 'testing', 'æ¸¬è©¦', 'é©—è­‰'],
            'è¨­è¨ˆ': ['design', 'architecture', 'è¨­è¨ˆ', 'æ¶æ§‹']
        }
        
        organized_files = []
        
        for file_path in source_path.rglob('*'):
            if file_path.is_file():
                filename = file_path.name.lower()
                
                # åˆ¤æ–·æ–‡ä»¶é¡å‹
                file_category = 'å…¶ä»–'
                for category, keywords in classification_rules.items():
                    if any(keyword in filename for keyword in keywords):
                        file_category = category
                        break
                
                # ç§»å‹•åˆ°å°æ‡‰ç›®éŒ„
                target_dir = self.kb_path / '01_raw' / file_category
                target_dir.mkdir(exist_ok=True)
                
                try:
                    target_path = target_dir / file_path.name
                    file_path.rename(target_path)
                    organized_files.append({
                        'original': str(file_path),
                        'new_location': str(target_path),
                        'category': file_category
                    })
                    print(f"   ğŸ“ {file_path.name} â†’ {file_category}/")
                except Exception as e:
                    print(f"   âŒ ç§»å‹•å¤±æ•— {file_path.name}: {e}")
        
        print(f"âœ… å®Œæˆæ–‡ä»¶æ•´ç†ï¼Œè™•ç† {len(organized_files)} å€‹æ–‡ä»¶")
        return organized_files
    
    def auto_tag_generation(self, tech_card_content):
        """
        2. è‡ªå‹•æ¨™ç±¤ç”Ÿæˆï¼šåŸºæ–¼å…§å®¹é—œéµå­—è‡ªå‹•å»ºè­°æ¨™ç±¤
        """
        print("ğŸ·ï¸ åŸ·è¡Œè‡ªå‹•æ¨™ç±¤ç”Ÿæˆ...")
        
        # é—œéµå­—åˆ°Domainçš„æ˜ å°„
        domain_keywords = {
            'Hardware': ['ç¡¬é«”', 'CPU', 'GPU', 'é¢¨æ‰‡', 'æ•£ç†±', 'é›»è·¯', 'ä¸»æ©Ÿæ¿', 'Server', 'æº«åº¦', 'PWM'],
            'Software': ['è»Ÿé«”', 'API', 'è³‡æ–™åº«', 'ç¨‹å¼', 'ç³»çµ±', 'ä»‹é¢', 'å¾Œç«¯', 'å‰ç«¯', 'bug', 'code'],
            'Process': ['æµç¨‹', 'ç®¡ç†', 'å®¢æˆ¶', 'æœå‹™', 'å“è³ª', 'SOP', 'åˆ¶åº¦', 'æ”¿ç­–', 'æœƒè­°', 'åŸ¹è¨“']
        }
        
        type_keywords = {
            'TechCard': ['æ–¹æ¡ˆ', 'è§£æ±º', 'æŠ€è¡“', 'å¯¦ä½œ', 'è¨­è¨ˆ', 'å„ªåŒ–', 'æ”¹å–„'],
            'ProcessCard': ['æµç¨‹', 'æ­¥é©Ÿ', 'ç¨‹åº', 'æŒ‡å—', 'è¦ç¯„', 'SOP'],
            'LessonCard': ['éŒ¯èª¤', 'å¤±æ•—', 'ç¶“é©—', 'æ•™è¨“', 'é¿å…', 'å•é¡Œ', 'é™·é˜±'],
            'RefCard': ['åƒè€ƒ', 'æ¨™æº–', 'è¦æ ¼', 'å°ç…§', 'æŸ¥è¡¨', 'ç¯„æœ¬']
        }
        
        # åˆ†ææ–‡æœ¬å…§å®¹
        text = (tech_card_content.get('title', '') + ' ' + 
               tech_card_content.get('problem', '') + ' ' +
               tech_card_content.get('solution', '')).lower()
        
        # æ¨è–¦Domain
        domain_scores = {}
        for domain, keywords in domain_keywords.items():
            score = sum(1 for keyword in keywords if keyword.lower() in text)
            domain_scores[domain] = score
        
        recommended_domain = max(domain_scores, key=domain_scores.get) if max(domain_scores.values()) > 0 else 'Process'
        
        # æ¨è–¦Type
        type_scores = {}
        for doc_type, keywords in type_keywords.items():
            score = sum(1 for keyword in keywords if keyword.lower() in text)
            type_scores[doc_type] = score
        
        recommended_type = max(type_scores, key=type_scores.get) if max(type_scores.values()) > 0 else 'TechCard'
        
        # æ¨è–¦æ©Ÿå¯†ç­‰ç´šï¼ˆç°¡åŒ–åˆ¤æ–·ï¼‰
        confidentiality_indicators = {
            'Restricted': ['å®¢æˆ¶', 'æ©Ÿå¯†', 'æˆæœ¬', 'åƒ¹æ ¼', 'åˆç´„'],
            'Dept': ['å…§éƒ¨', 'éƒ¨é–€', 'åœ˜éšŠ'],
            'Public': ['ä¸€èˆ¬', 'é€šç”¨', 'æ¨™æº–']
        }
        
        recommended_confidentiality = 'Public'  # é»˜èª
        for level, indicators in confidentiality_indicators.items():
            if any(indicator in text for indicator in indicators):
                recommended_confidentiality = level
                break
        
        suggested_tags = f"#{recommended_domain} #{recommended_type} #{recommended_confidentiality}"
        
        # ä¿¡å¿ƒåº¦è¨ˆç®—
        confidence = {
            'domain': min(domain_scores[recommended_domain] / 3, 1.0),
            'type': min(type_scores[recommended_type] / 2, 1.0),
            'overall': (domain_scores[recommended_domain] + type_scores[recommended_type]) / 5
        }
        
        print(f"   ğŸ¯ å»ºè­°æ¨™ç±¤ï¼š{suggested_tags}")
        print(f"   ğŸ“Š ä¿¡å¿ƒåº¦ï¼š{confidence['overall']:.1%}")
        
        return {
            'suggested_tags': suggested_tags,
            'domain': recommended_domain,
            'type': recommended_type,
            'confidentiality': recommended_confidentiality,
            'confidence': confidence
        }
    
    def auto_format_generation(self, tech_card, tags_info):
        """
        3. è‡ªå‹•æ ¼å¼ç”Ÿæˆï¼šç”Ÿæˆæ¨™æº–Markdownæ ¼å¼
        """
        print("ğŸ“„ åŸ·è¡Œè‡ªå‹•æ ¼å¼ç”Ÿæˆ...")
        
        today = datetime.now().strftime("%Y-%m-%d")
        card_id = f"TC-{today.replace('-', '')}-{hash(tech_card['title']) % 1000:03d}"
        
        markdown_content = f"""
# æŠ€è¡“å¡ï¼š{tech_card['title']}

## ğŸ“‹ åŸºæœ¬è³‡è¨Š
- **æŠ€è¡“å¡ID**: {card_id}
- **å»ºç«‹æ—¥æœŸ**: {today}
- **å»ºç«‹è€…**: [è‡ªå‹•ç”Ÿæˆ]
- **åˆ†é¡æ¨™ç±¤**: {tags_info['suggested_tags']}
- **ä¿¡å¿ƒåº¦**: {tags_info['confidence']['overall']:.1%}

## â“ å•é¡Œæè¿°
{tech_card.get('problem', '').strip()}

## ğŸ’¡ è§£æ±ºæ–¹æ¡ˆ
{tech_card.get('solution', '').strip()}

## ğŸ“Š é—œéµåƒæ•¸
{tech_card.get('parameters', '').strip()}

## ğŸ¯ é©ç”¨ç¯„åœèˆ‡é™åˆ¶
{tech_card.get('scope', '').strip()}

## ğŸ”— ç›¸é—œè³‡æº
- **ä¾†æºæ–‡ä»¶**: [å¾…è£œå……]
- **è¯çµ¡äºº**: [å¾…è£œå……]
- **æ›´æ–°è¨˜éŒ„**: {today} - è‡ªå‹•ç”Ÿæˆåˆç‰ˆ

---
*ğŸ¤– æœ¬æ–‡æª”ç”±çŸ¥è­˜æ²»ç†è‡ªå‹•åŒ–å·¥å…·ç”Ÿæˆï¼Œè«‹å¯©æ ¸å¾Œç™¼å¸ƒ*
""".strip()
        
        # ä¿å­˜åˆ°æ–‡ä»¶
        output_dir = self.kb_path / '02_processed'
        output_file = output_dir / f"{card_id}.md"
        
        with open(output_file, 'w', encoding='utf-8') as f:
            f.write(markdown_content)
        
        print(f"   âœ… å·²ç”Ÿæˆï¼š{output_file}")
        
        return {
            'card_id': card_id,
            'file_path': str(output_file),
            'markdown_content': markdown_content
        }
    
    def auto_index_update(self, processed_cards):
        """
        4. è‡ªå‹•ç´¢å¼•æ›´æ–°ï¼šæ›´æ–°æœå°‹ç´¢å¼•å’Œçµ±è¨ˆè³‡æ–™
        """
        print("ğŸ” åŸ·è¡Œè‡ªå‹•ç´¢å¼•æ›´æ–°...")
        
        # å»ºç«‹æœå°‹ç´¢å¼•
        search_index = []
        category_stats = {}
        
        for card in processed_cards:
            # æœå°‹ç´¢å¼•é …ç›®
            index_entry = {
                'id': card.get('card_id', ''),
                'title': card.get('title', ''),
                'keywords': self._extract_keywords(card),
                'tags': card.get('tags', ''),
                'summary': card.get('problem', '')[:100] + '...',
                'created_date': datetime.now().strftime("%Y-%m-%d")
            }
            search_index.append(index_entry)
            
            # åˆ†é¡çµ±è¨ˆ
            tags = card.get('tags', '')
            if tags:
                for tag in tags.split():
                    tag = tag.replace('#', '')
                    category_stats[tag] = category_stats.get(tag, 0) + 1
        
        # ä¿å­˜ç´¢å¼•æ–‡ä»¶
        index_file = self.kb_path / '03_indexed' / 'search_index.json'
        with open(index_file, 'w', encoding='utf-8') as f:
            json.dump(search_index, f, ensure_ascii=False, indent=2)
        
        # ä¿å­˜çµ±è¨ˆæ–‡ä»¶
        stats_file = self.kb_path / '04_metadata' / 'category_stats.json'
        with open(stats_file, 'w', encoding='utf-8') as f:
            json.dump(category_stats, f, ensure_ascii=False, indent=2)
        
        print(f"   ğŸ“Š ç´¢å¼•æ›´æ–°å®Œæˆï¼š{len(search_index)} å€‹é …ç›®")
        print(f"   ğŸ“ˆ åˆ†é¡çµ±è¨ˆï¼š{len(category_stats)} å€‹åˆ†é¡")
        
        return {
            'search_index': search_index,
            'category_stats': category_stats,
            'index_file': str(index_file)
        }
    
    def _extract_keywords(self, card):
        """å¾æŠ€è¡“å¡ä¸­æå–é—œéµå­—"""
        text = ' '.join([
            card.get('title', ''),
            card.get('problem', ''),
            card.get('solution', '')
        ])
        
        # ç°¡å–®çš„é—œéµå­—æå–
        words = re.findall(r'\b[a-zA-Z\u4e00-\u9fff]{3,}\b', text)
        return list(set(words))[:10]  # å–å‰10å€‹ä¸é‡è¤‡çš„é—œéµå­—

# ç¤ºç¯„ Phase 1 å·¥å…·ä½¿ç”¨
print("ğŸ› ï¸ Phase 1 è‡ªå‹•åŒ–å·¥å…·æ¼”ç¤º")
print("=" * 50)

# åˆå§‹åŒ–å·¥å…·
automation_tools = BasicAutomationTools()

# ç¤ºç¯„æŠ€è¡“å¡
sample_card = {
    'title': 'APIå›æ‡‰æ™‚é–“å„ªåŒ–æ–¹æ¡ˆ',
    'problem': 'ç³»çµ±APIå›æ‡‰æ™‚é–“éé•·ï¼Œå½±éŸ¿ç”¨æˆ¶é«”é©—',
    'solution': '1. å¢åŠ è³‡æ–™åº«ç´¢å¼• 2. å¯¦ä½œå¿«å–æ©Ÿåˆ¶ 3. å„ªåŒ–æŸ¥è©¢èªå¥',
    'parameters': 'Rediså¿«å–ï¼ŒTTL=300ç§’ï¼Œç´¢å¼•è¦†è“‹ç‡90%',
    'scope': 'é©ç”¨æ–¼æŸ¥è©¢é »ç¹çš„APIï¼Œä¸é©ç”¨æ–¼å¯¦æ™‚æ€§è¦æ±‚æ¥µé«˜çš„å ´æ™¯'
}

# åŸ·è¡Œè‡ªå‹•åŒ–æµç¨‹
tags_info = automation_tools.auto_tag_generation(sample_card)
formatted_card = automation_tools.auto_format_generation(sample_card, tags_info)
index_result = automation_tools.auto_index_update([{**sample_card, **formatted_card, 'tags': tags_info['suggested_tags']}])

print("\nâœ… Phase 1 è‡ªå‹•åŒ–å·¥å…·æ¼”ç¤ºå®Œæˆ")
print(f"ğŸ“Š é ä¼°ç¯€çœæ™‚é–“ï¼šæ¯å¼µæŠ€è¡“å¡ç´„10-15åˆ†é˜")
print(f"ğŸ¯ æŠ•è³‡å›å ±ï¼šå·¥å…·é–‹ç™¼2å¤©ï¼Œæ¯æœˆå¯ç¯€çœ4-6å°æ™‚")
````

````python
# ğŸš€ Phase 2: æ™ºèƒ½åŒ–è‡ªå‹•åŒ–å·¥å…·

import re
from collections import Counter

class SmartAutomationTools:
    """
    Phase 2 æ™ºèƒ½åŒ–è‡ªå‹•åŒ–å·¥å…·
    å¼•å…¥æ›´å¤šAIå’Œæ©Ÿå™¨å­¸ç¿’èƒ½åŠ›
    """
    
    def __init__(self):
        self.document_templates = self._load_templates()
        self.quality_model = self._init_quality_model()
    
    def smart_content_extraction(self, document_path):
        """
        æ™ºèƒ½å…§å®¹æå–ï¼šå¾æ–‡æª”ä¸­è‡ªå‹•æå–æŠ€è¡“å¡è¦ç´ 
        """
        print("ğŸ§  åŸ·è¡Œæ™ºèƒ½å…§å®¹æå–...")
        
        # æ¨¡æ“¬è®€å–æ–‡æª”ï¼ˆå¯¦éš›æ‡‰è©²ç”¨OCRæˆ–æ–‡æª”è§£æï¼‰
        sample_document_content = """
        å°ˆæ¡ˆåç¨±ï¼šServeræ•£ç†±æ”¹é€²å°ˆæ¡ˆ
        
        å•é¡Œæè¿°ï¼š
        å·¥å» ç’°å¢ƒä¸­çš„Serverä¸»æ©Ÿç¶“å¸¸å‡ºç¾éç†±å•é¡Œï¼ŒCPUæº«åº¦é”åˆ°75-80Â°Cï¼Œ
        å°è‡´ç³»çµ±è‡ªå‹•é™é »ï¼Œå½±éŸ¿ç”Ÿç”¢ç·šæ•¸æ“šè™•ç†æ•ˆèƒ½ã€‚
        
        è§£æ±ºæ–¹æ¡ˆï¼š
        1. æ›´æ›ç‚ºPWMèª¿é€Ÿé¢¨æ‰‡
        2. é‡æ–°è¨­è¨ˆæ•£ç†±è·¯å¾‘
        3. åŠ è£æº«åº¦ç›£æ§ç³»çµ±
        
        å¯¦æ–½çµæœï¼š
        CPUæº«åº¦é™è‡³55-60Â°Cï¼Œç³»çµ±ç©©å®šé‹è¡Œï¼Œæ€§èƒ½æå‡25%ã€‚
        æˆæœ¬ï¼šç¡¬é«”$150ï¼Œäººå·¥$300ï¼Œç¸½è¨ˆ$450ã€‚
        
        é©ç”¨ç¯„åœï¼š
        é©ç”¨æ–¼å·¥æ¥­ç’°å¢ƒServerï¼Œä¸é©ç”¨æ–¼ç„¡å¡µç’°å¢ƒã€‚
        """
        
        # ä½¿ç”¨æ­£å‰‡è¡¨é”å¼å’ŒNLPæŠ€è¡“æå–é—œéµè³‡è¨Š
        extracted_data = self._extract_structured_content(sample_document_content)
        
        print(f"   âœ… æå–å®Œæˆï¼Œä¿¡å¿ƒåº¦ï¼š{extracted_data['confidence']:.1%}")
        
        return extracted_data
    
    def _extract_structured_content(self, text):
        """å¾éçµæ§‹åŒ–æ–‡æœ¬ä¸­æå–çµæ§‹åŒ–å…§å®¹"""
        
        # å®šç¾©åŒ¹é…æ¨¡å¼
        patterns = {
            'title': r'å°ˆæ¡ˆåç¨±[ï¼š:](.*)',
            'problem': r'å•é¡Œæè¿°[ï¼š:](.*?)(?=è§£æ±ºæ–¹æ¡ˆ|$)',
            'solution': r'è§£æ±ºæ–¹æ¡ˆ[ï¼š:](.*?)(?=å¯¦æ–½çµæœ|é©ç”¨ç¯„åœ|$)',
            'results': r'å¯¦æ–½çµæœ[ï¼š:](.*?)(?=é©ç”¨ç¯„åœ|$)',
            'scope': r'é©ç”¨ç¯„åœ[ï¼š:](.*?)(?=$)'
        }
        
        extracted = {}
        confidence_scores = []
        
        for field, pattern in patterns.items():
            match = re.search(pattern, text, re.DOTALL | re.IGNORECASE)
            if match:
                content = match.group(1).strip()
                extracted[field] = content
                confidence_scores.append(0.9)  # åŒ¹é…åˆ°å‰‡é«˜ä¿¡å¿ƒåº¦
            else:
                extracted[field] = ""
                confidence_scores.append(0.1)  # æœªåŒ¹é…å‰‡ä½ä¿¡å¿ƒåº¦
        
        # å¾Œè™•ç†ï¼šæå–æ•¸å€¼å’Œé—œéµåƒæ•¸
        parameters = self._extract_parameters(text)
        extracted['parameters'] = parameters
        
        # é‡æ–°çµ„ç¹”ç‚ºæŠ€è¡“å¡æ ¼å¼
        tech_card = {
            'title': extracted['title'] or 'è‡ªå‹•æå–çš„æŠ€è¡“å¡',
            'problem': extracted['problem'],
            'solution': extracted['solution'],
            'parameters': extracted['parameters'],
            'scope': extracted['scope']
        }
        
        overall_confidence = sum(confidence_scores) / len(confidence_scores)
        
        return {
            'tech_card': tech_card,
            'confidence': overall_confidence,
            'extraction_details': extracted
        }
    
    def _extract_parameters(self, text):
        """æå–é—œéµåƒæ•¸å’Œæ•¸å€¼"""
        parameters = []
        
        # æº«åº¦æ•¸æ“š
        temp_matches = re.findall(r'(\d+)[-~](\d+)?Â°C', text)
        for match in temp_matches:
            if match[1]:  # ç¯„åœ
                parameters.append(f"æº«åº¦ç¯„åœï¼š{match[0]}-{match[1]}Â°C")
            else:  # å–®ä¸€å€¼
                parameters.append(f"æº«åº¦ï¼š{match[0]}Â°C")
        
        # æˆæœ¬æ•¸æ“š
        cost_matches = re.findall(r'\$(\d+)', text)
        if cost_matches:
            total_cost = sum(int(cost) for cost in cost_matches)
            parameters.append(f"ç¸½æˆæœ¬ï¼š${total_cost}")
        
        # æ€§èƒ½æå‡
        perf_matches = re.findall(r'æå‡(\d+)%', text)
        for match in perf_matches:
            parameters.append(f"æ€§èƒ½æå‡ï¼š{match}%")
        
        return '\n'.join(parameters) if parameters else 'ç„¡ç‰¹å®šæ•¸å€¼åƒæ•¸'
    
    def smart_quality_prediction(self, tech_card):
        """
        æ™ºèƒ½å“è³ªé æ¸¬ï¼šåŸºæ–¼æ­·å²è³‡æ–™é æ¸¬æŠ€è¡“å¡å“è³ªåˆ†æ•¸
        """
        print("ğŸ“Š åŸ·è¡Œæ™ºèƒ½å“è³ªé æ¸¬...")
        
        # åŸºæ–¼è¦å‰‡çš„å“è³ªè©•åˆ†ï¼ˆç°¡åŒ–ç‰ˆæ©Ÿå™¨å­¸ç¿’æ¨¡å‹ï¼‰
        features = self._extract_quality_features(tech_card)
        predicted_score = self._calculate_quality_score(features)
        
        # æä¾›æ”¹é€²å»ºè­°
        suggestions = self._generate_improvement_suggestions(features, predicted_score)
        
        print(f"   ğŸ¯ é æ¸¬å“è³ªï¼š{predicted_score:.2f}/5.0")
        print(f"   ğŸ’¡ æ”¹é€²å»ºè­°ï¼š{len(suggestions)} é …")
        
        return {
            'predicted_score': predicted_score,
            'features': features,
            'suggestions': suggestions
        }
    
    def _extract_quality_features(self, tech_card):
        """æå–å“è³ªè©•ä¼°ç‰¹å¾µ"""
        features = {}
        
        # é•·åº¦ç‰¹å¾µ
        for field in ['problem', 'solution', 'parameters', 'scope']:
            content = tech_card.get(field, '')
            features[f'{field}_length'] = len(content.split())
            features[f'{field}_has_content'] = len(content.strip()) > 0
        
        # çµæ§‹ç‰¹å¾µ
        solution = tech_card.get('solution', '')
        features['has_numbered_steps'] = bool(re.search(r'[1-9]\.|ç¬¬[ä¸€äºŒä¸‰å››äº”]', solution))
        features['has_bullet_points'] = 'â€¢' in solution or '-' in solution
        
        # æ•¸æ“šç‰¹å¾µ
        params = tech_card.get('parameters', '')
        features['has_numbers'] = bool(re.search(r'\d+', params))
        features['has_units'] = bool(re.search(r'%|Â°C|dB|RPM|\$|MB|GB', params))
        
        # ç¯„åœç‰¹å¾µ
        scope = tech_card.get('scope', '')
        features['defines_limits'] = any(word in scope for word in ['ä¸é©ç”¨', 'é™åˆ¶', 'åƒ…é™', 'åªèƒ½'])
        features['defines_conditions'] = any(word in scope for word in ['é©ç”¨', 'æ¢ä»¶', 'ç’°å¢ƒ', 'æƒ…æ³'])
        
        return features
    
    def _calculate_quality_score(self, features):
        """åŸºæ–¼ç‰¹å¾µè¨ˆç®—å“è³ªåˆ†æ•¸"""
        score = 2.5  # åŸºæº–åˆ†æ•¸
        
        # å®Œæ•´æ€§è©•åˆ†
        completeness = sum(features[f'{field}_has_content'] for field in ['problem', 'solution', 'parameters', 'scope'])
        score += completeness * 0.3
        
        # çµæ§‹æ¸…æ™°åº¦
        if features['has_numbered_steps'] or features['has_bullet_points']:
            score += 0.4
        
        # æ•¸æ“šè±å¯Œåº¦
        if features['has_numbers'] and features['has_units']:
            score += 0.3
        
        # ç¯„åœæ˜ç¢ºåº¦
        if features['defines_limits'] and features['defines_conditions']:
            score += 0.3
        
        # é•·åº¦é©ä¸­æ€§ï¼ˆé¿å…éçŸ­æˆ–éé•·ï¼‰
        total_words = sum(features[f'{field}_length'] for field in ['problem', 'solution', 'parameters', 'scope'])
        if 50 <= total_words <= 300:
            score += 0.2
        elif total_words < 30:
            score -= 0.5
        
        return min(max(score, 1.0), 5.0)
    
    def _generate_improvement_suggestions(self, features, score):
        """ç”Ÿæˆæ”¹é€²å»ºè­°"""
        suggestions = []
        
        if not features['problem_has_content']:
            suggestions.append("è£œå……å•é¡Œæè¿°ï¼Œèªªæ˜å…·é«”ç—‡ç‹€å’Œå½±éŸ¿")
        
        if not features['has_numbered_steps'] and not features['has_bullet_points']:
            suggestions.append("ç‚ºè§£æ±ºæ–¹æ¡ˆå¢åŠ ç·¨è™Ÿæˆ–åˆ†é»ï¼Œæå‡å¯è®€æ€§")
        
        if not features['has_numbers']:
            suggestions.append("è£œå……å…·é«”çš„æ•¸å€¼åƒæ•¸ï¼Œå¢å¼·å¯ä¿¡åº¦")
        
        if not features['defines_limits']:
            suggestions.append("æ˜ç¢ºèªªæ˜é©ç”¨é™åˆ¶ï¼Œé¿å…èª¤ç”¨")
        
        if score < 3.5:
            suggestions.append("å»ºè­°è«‹åŒäº‹å¯©æ ¸ï¼Œæå‡å…§å®¹å“è³ª")
        
        return suggestions
    
    def auto_monthly_report(self, knowledge_base_path):
        """
        è‡ªå‹•æœˆå ±ç”Ÿæˆï¼šåˆ†æçŸ¥è­˜åº«æ•¸æ“šç”Ÿæˆæœˆåº¦å ±å‘Š
        """
        print("ğŸ“Š åŸ·è¡Œè‡ªå‹•æœˆå ±ç”Ÿæˆ...")
        
        # æ¨¡æ“¬çŸ¥è­˜åº«æ•¸æ“šåˆ†æ
        mock_data = {
            'total_cards': 45,
            'new_cards_this_month': 12,
            'average_quality': 4.1,
            'search_queries': 156,
            'top_categories': ['Hardware', 'Software', 'Process'],
            'most_viewed_cards': ['PWMé¢¨æ‰‡æ§åˆ¶', 'APIå„ªåŒ–æ–¹æ¡ˆ', 'Bugä¿®å¾©æµç¨‹']
        }
        
        report_content = self._generate_report_content(mock_data)
        
        # ä¿å­˜å ±å‘Š
        report_file = f"æœˆåº¦çŸ¥è­˜æ²»ç†å ±å‘Š_{datetime.now().strftime('%Y-%m')}.md"
        
        print(f"   ğŸ“„ å ±å‘Šå·²ç”Ÿæˆï¼š{report_file}")
        
        return {
            'report_content': report_content,
            'report_file': report_file,
            'statistics': mock_data
        }
    
    def _generate_report_content(self, data):
        """ç”Ÿæˆå ±å‘Šå…§å®¹"""
        report = f"""
# ğŸ“Š æœˆåº¦çŸ¥è­˜æ²»ç†å ±å‘Š - {datetime.now().strftime('%Yå¹´%mæœˆ')}

## ğŸ“ˆ ç¸½é«”è¡¨ç¾
- **çŸ¥è­˜åº«è¦æ¨¡**: {data['total_cards']} å¼µæŠ€è¡“å¡
- **æœ¬æœˆæ–°å¢**: {data['new_cards_this_month']} å¼µ
- **å¹³å‡å“è³ª**: {data['average_quality']}/5.0
- **ä½¿ç”¨é »æ¬¡**: {data['search_queries']} æ¬¡æœå°‹

## ğŸ† æœ¬æœˆäº®é»
- æ–°å¢æŠ€è¡“å¡æ•¸é‡é”åˆ°ç›®æ¨™
- å¹³å‡å“è³ªåˆ†æ•¸è¶…é4.0
- ä½¿ç”¨é »æ¬¡ç©©å®šæˆé•·

## ğŸ“Š åˆ†é¡åˆ†ä½ˆ
æ’åå‰ä¸‰çš„åˆ†é¡ï¼š
1. {data['top_categories'][0]}
2. {data['top_categories'][1]}
3. {data['top_categories'][2]}

## ğŸ”¥ ç†±é–€å…§å®¹
æœ€å—æ­¡è¿çš„æŠ€è¡“å¡ï¼š
1. {data['most_viewed_cards'][0]}
2. {data['most_viewed_cards'][1]}
3. {data['most_viewed_cards'][2]}

## ğŸ¯ ä¸‹æœˆç›®æ¨™
- æ–°å¢æŠ€è¡“å¡ç›®æ¨™ï¼š15å¼µ
- å“è³ªåˆ†æ•¸ç›®æ¨™ï¼š4.3/5.0
- ä½¿ç”¨ç‡æå‡ï¼š20%

---
*ğŸ¤– æœ¬å ±å‘Šç”±è‡ªå‹•åŒ–ç³»çµ±ç”Ÿæˆæ–¼ {datetime.now().strftime('%Y-%m-%d %H:%M')}*
""".strip()
        return report
    
    def _load_templates(self):
        """è¼‰å…¥æ–‡æª”æ¨¡æ¿"""
        return {
            'tech_card': 'å››è¦ç´ æŠ€è¡“å¡æ¨¡æ¿',
            'process_card': 'æµç¨‹å¡æ¨¡æ¿',
            'lesson_card': 'ç¶“é©—æ•™è¨“å¡æ¨¡æ¿'
        }
    
    def _init_quality_model(self):
        """åˆå§‹åŒ–å“è³ªè©•ä¼°æ¨¡å‹"""
        return {'model_version': '2.0', 'accuracy': 0.85}

# ç¤ºç¯„ Phase 2 å·¥å…·ä½¿ç”¨
print("ğŸ§  Phase 2 æ™ºèƒ½åŒ–å·¥å…·æ¼”ç¤º")
print("=" * 50)

smart_tools = SmartAutomationTools()

# æ¼”ç¤ºæ™ºèƒ½å…§å®¹æå–
extraction_result = smart_tools.smart_content_extraction("sample_document.pdf")
print(f"\nğŸ“„ æå–çš„æŠ€è¡“å¡æ¨™é¡Œï¼š{extraction_result['tech_card']['title']}")

# æ¼”ç¤ºæ™ºèƒ½å“è³ªé æ¸¬
quality_prediction = smart_tools.smart_quality_prediction(extraction_result['tech_card'])
print(f"\nğŸ“Š å“è³ªé æ¸¬çµæœï¼š{quality_prediction['predicted_score']:.2f}/5.0")

# æ¼”ç¤ºè‡ªå‹•æœˆå ±ç”Ÿæˆ
monthly_report = smart_tools.auto_monthly_report("./knowledge_base")
print(f"\nğŸ“‹ æœˆå ±å·²ç”Ÿæˆï¼š{monthly_report['report_file']}")

print("\nâœ… Phase 2 æ™ºèƒ½åŒ–å·¥å…·æ¼”ç¤ºå®Œæˆ")
print(f"ğŸ“Š é ä¼°ç¯€çœæ™‚é–“ï¼šæ¯æœˆé¡å¤–ç¯€çœ8-12å°æ™‚")
print(f"ğŸ¯ æŠ•è³‡å›å ±ï¼šå·¥å…·é–‹ç™¼1é€±ï¼ŒROIåœ¨3å€‹æœˆå…§é¡¯ç¾")
````
