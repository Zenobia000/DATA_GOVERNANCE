## 🔍 三問法：判斷文件價值的黃金標準

每當你拿到一份文件（專案報告、會議紀錄、技術文件），先問自己這三個問題：

### ❓ 問題1：其他專案會用到嗎？
**判斷標準**：
- ✅ **高價值**：包含通用規則、標準流程、設計原則
- ❌ **低價值**：只針對特定客戶、特定時間的個案性內容

**範例**：
```
✅ 好例子：「PWM風扇控制設計規範」（其他硬體專案都能參考）
❌ 壞例子：「ABC公司2025年1月會議紀錄」（太具體，難以複用）
```

### ❓ 問題2：未來一年內會重複遇到嗎？
**判斷標準**：
- ✅ **高價值**：解決的是經常發生的問題
- ❌ **低價值**：解決的是一次性、臨時性問題

**範例**：
```
✅ 好例子：「客戶投訴處理SOP」（客服問題經常發生）
❌ 壞例子：「2025年春節放假通知」（只用一次）
```

### ❓ 問題3：不存會不會「再次踩雷」？
**判斷標準**：
- ✅ **高價值**：包含重要教訓、避坑指南、失敗經驗
- ❌ **低價值**：沒有特殊學習價值的一般性記錄

**範例**：
```
✅ 好例子：「散熱設計失敗案例分析」（避免重複犯錯）
❌ 壞例子：「例行週報範本」（沒有特殊技巧）
```

## ⚡ 評分規則：簡單粗暴

**評分方式**：
- 每個問題答「是」= 1分
- 每個問題答「否」= 0分

**決策標準**：
- **3分**：🌟 **必存** - 立即進知識庫
- **2分**：⚠️ **考慮** - 看資源決定
- **1分**：❓ **觀察** - 暫時不存，但記錄原因
- **0分**：🗑️ **丟掉** - 直接刪除或歸檔

### 🧮 自動計算工具

如果你想要程式化評估，可以使用下面的程式碼：

````python
# 文件價值評估工具
def evaluate_document(filename, q1_reusable, q2_recurring, q3_lessons):
    """
    文件價值評估函數
    
    參數:
        filename: 文件名稱
        q1_reusable: 其他專案會用到嗎？ (True/False)
        q2_recurring: 一年內會重複遇到嗎？ (True/False) 
        q3_lessons: 不存會踩雷嗎？ (True/False)
    
    返回:
        評估結果字典
    """
    score = sum([q1_reusable, q2_recurring, q3_lessons])
    
    if score == 3:
        decision = "🌟 必存"
        color = "green"
    elif score == 2:
        decision = "⚠️ 考慮"
        color = "orange"
    elif score == 1:
        decision = "❓ 觀察"
        color = "yellow"
    else:
        decision = "🗑️ 丟掉"
        color = "red"
    
    return {
        'filename': filename,
        'q1': '✅' if q1_reusable else '❌',
        'q2': '✅' if q2_recurring else '❌', 
        'q3': '✅' if q3_lessons else '❌',
        'score': score,
        'decision': decision,
        'color': color
    }

# 範例評估
examples = [
    ("PWM風扇控制方案.pdf", True, True, True),
    ("ABC公司會議紀錄.docx", False, False, False),
    ("客戶投訴處理SOP.md", True, True, True),
    ("2025春節放假通知.pdf", False, False, False),
    ("散熱設計失敗案例.pptx", True, True, True)
]

print("📊 文件價值評估結果")
print("=" * 80)
print(f"{'文件名稱':<25} {'Q1':<4} {'Q2':<4} {'Q3':<4} {'分數':<4} {'決策':<10}")
print("-" * 80)

for filename, q1, q2, q3 in examples:
    result = evaluate_document(filename, q1, q2, q3)
    print(f"{result['filename']:<25} {result['q1']:<4} {result['q2']:<4} {result['q3']:<4} {result['score']:<4} {result['decision']:<10}")

print("\n💡 學習重點：")
print("1. 高價值文件通常同時滿足『可複用』、『會重複』、『有教訓』三個條件")
print("2. 一次性、個案性的文件即使內容豐富，價值也相對較低")
print("3. 寧可錯殺也不要錯放 - 知識庫品質比數量更重要")
````
